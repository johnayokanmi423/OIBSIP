### Histogram plot for the mean of Menu

## Comparative Analysis

```python

#Select categories by custom count
top_categories=df['Category'].value_counts().head(5).index
df_top=df[df['Category'].isin(top_categories)]
#Create a 2,2 plot
fig,axes=plt.subplots(2,2, figsize=(16,12))

#Strip plot 
sns.stripplot(data=df_top, x='Category', y='Calories', size=8, alpha=0.7, ax=axes[0,0])
axes[0,0].set_title('Calories Distribution by category')
axes[0,0].tick_params(axis='x', rotation=45)

#Swarm plot
sns.swarmplot(data=df_top, x='Category', y='Saturated Fat', size=8, alpha=0.7, ax=axes[0,1])
axes[0,1].set_title('Saturated fat Distribution by category')
axes[0,1].tick_params(axis='x', rotation=45)

#Sugars
sns.stripplot(data=df_top, x='Category', y='Sugars', size=8, alpha=0.7, ax=axes[1,0])
axes[1,0].set_title('Saturated fat Distribution by category')
axes[1,0].tick_params(axis='x', rotation=45)

#Sodium
sns.swarmplot(data=df_top, x='Category', y='Sodium', size=8, alpha=0.7, ax=axes[1,1])
axes[1,1].set_title('Saturated fat Distribution by category')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## Dashboard Overview

```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd

category_counts = df['Category'].value_counts()

#Create a 3*4 grid
fig=plt.figure(figsize=(40,12))
gs=fig.add_gridspec(2,4, hspace=0.3, wspace=0.3)
fig.suptitle("Dashboard Overview Of The Menu", fontsize=15, fontweight='bold')

#Category distribution
ax1=fig.add_subplot(gs[0,0])
category_counts.plot(kind='bar', ax=ax1, color=sns.color_palette('husl', len(category_counts)))
ax1.set_title('Items by category', fontweight= 'bold')
ax1.tick_params(axis='x', rotation=45)

#Calories Distribution
ax2=fig.add_subplot(gs[0,1])
sns.distplot(df['Calories'], kde= True, ax=ax2)
ax2.set_title('Calories distribution', fontweight= 'bold')

#Top 10 Calorie item
ax3=fig.add_subplot(gs[0,2:])
top_calorie_item=df.nlargest(10, 'Calories')
sns.barplot(data=top_calorie_item, ax=ax3, x='Calories', y='Item')
ax3.set_title('Top 10 Calorie Item', fontweight= 'bold')

#Nutritional Correlation
ax4= fig.add_subplot(gs[1,:2])
Nutritional_corr=df[['Sodium', 'Total Fat', 'Saturated Fat', 'Cholesterol']].corr()
sns.heatmap(Nutritional_corr, annot=True, ax=ax4, cmap='BrBG')
ax4.set_title('Nutritional correlation', fontweight='bold')

#Category wise Calories
ax5=fig.add_subplot(gs[1,2:])
sns.boxplot(data=df, x='Category', y='Calories', ax=ax5)
ax5.set_title("Category wise Calories", fontweight= 'bold')
ax5.tick_params(axis='x', rotation=45)
```

## Discussion
1. The largest item category is coffee & Tea which means that the customers order that more when compared to ther products
2. The Chicken McNuggets (40 piece) contributes more calorie than any other item found on the list
3. There is a high correlation value of 0.85 between the Total fat and sodium, and the same value is repeated in the case of total fat and saturated fat. This show a form of relationship between the two Nutritional criteria.
4. The average quantity of sald was about 4.5 which is high compared to other items on the menu, indicating that people order a very high quantity of salad compared to other items.

## Recommendation
1. An awareness campaign should be created to enable customers understand while it is important to explore other items on the menu. Factors like nutritional contribution of each item on the menu should be the main highlight.
2. If the aim is to drive more revenue based on the Analysis. Then it will be better to focus on the most performing item on the menu. The Chicken McNuggets (40 Piece) is the item that contributes the most to the total revenue generated by the Store.
3. Since there is high correlation of 0.85 between the Total fat, Sodium, and Total Saturated fat. The items can be comsidered as supplements especially for customers who are concerned about nutrtional values more than the specific item on the menu.
